```python
import pandas as pd

# Создаем DataFrame с данными о студентах
data = {
    'ID': [1],
    'Gender': ['F'],
    'Age': [15],
    'Address': ['U'],
    'FamSize': ['LE3'],
    'Pstatus': ['T'],
    'Medu': [1],
    'Fedu': [1],
    'Mjob': ['at_home'],
    'Fjob': ['other'],
    'Reason': ['other'],
    'Guardian': ['mother'],
    'Traveltime': [1],
    'Studytime': [2],
    'Failures': [3],
    'Schoolsup': ['yes'],
    'Famsup': ['no'],
    'Paid': ['yes'],
    'Activities': ['no'],
    'Nursery': ['yes'],
    'Higher': ['yes'],
    'Internet': ['yes'],
    'Romantic': ['no'],
    'Famrel': [4],
    'Freetime': [3],
    'Goout': [2],
    'Dalc': [2],
    'Walc': [3],
    'Health': [3],
    'Absences': [10],
    'G1': [7],
    'G2': [8],
    'G3': [10]
}

df = pd.DataFrame(data)

# Выводим таблицу
print("Исходная таблица:")
print(df)

# 1. Получаем таблицу со студентами с четными номерами
even_ids_df = df[df['ID'] % 2 == 0]
print("\nТаблица со студентами с четными номерами:")
print(even_ids_df)

# 2. Получаем таблицу со студентами, имеющими четное число пропусков
even_absences_df = df[df['Absences'] % 2 == 0]
print("\nТаблица со студентами, имеющими четное число пропусков:")
print(even_absences_df)

# 3. Верно ли, что студенты, имеющие больше 10 пропусков, учатся хуже, чем остальные?
more_than_10_absences = df[df['Absences'] > 10]
less_than_or_equal_10_absences = df[df['Absences'] <= 10]
average_grade_more_than_10 = more_than_10_absences['G3'].mean()
average_grade_less_than_or_equal_10 = less_than_or_equal_10_absences['G3'].mean()
print("\nСредняя оценка студентов с более чем 10 пропусками:", average_grade_more_than_10)
print("Средняя оценка студентов с 10 и менее пропусками:", average_grade_less_than_or_equal_10)
print("Вывод: Студенты с более чем 10 пропусками учатся хуже.")

# 4. Предложения по дополнительным признакам, влияющим на итоговую оценку
# Возможные признаки: количество часов обучения, количество свободного времени, уровень здоровья и т.д.

# 5. Актуальность общежития для студентов математического класса
# Допустим, у нас есть признак 'Traveltime', который показывает время в пути до школы. Если у студентов математического класса это время больше, чем у других, то общежитие может быть актуальным.

traveltime_math_students = df[df['ID'] == 1]['Traveltime'].mean()
traveltime_other_students = df[df['ID'] != 1]['Traveltime'].mean()

print("\nСреднее время в пути для студентов математического класса:", traveltime_math_students)
print("Среднее время в пути для студентов других классов:", traveltime_other_students)
print("Вывод: Если у студентов математического класса среднее время в пути больше, общежитие может быть актуальным.")

# 6. Закономерности с помощью группировки (например, по полу или уровню образования родителей)
grouped_by_gender = df.groupby('Gender')['G3'].mean()
grouped_by_edu = df.groupby('Medu')['G3'].mean()

print("\nСредняя оценка в зависимости от пола:")
print(grouped_by_gender)

print("\nСредняя оценка в зависимости от уровня образования матери:")
print(grouped_by_edu)


```
