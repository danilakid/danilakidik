```python
import numpy as np
import matplotlib.pyplot as plt

# Генерация выборки из 1000 элементов с равномерным распределением от -1 до 7
data = np.random.uniform(-1, 7, 1000)

# Построение гистограммы
plt.hist(data, bins=20, color='blue', edgecolor='black', alpha=0.7)
plt.title('Histogram of Uniform Distribution')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Задаем диапазон значений x
x_values = np.linspace(-5, 12, 400)

# Вычисляем значения функций
y1_values = x_values**2
y2_values = 10 * np.sin(x_values)
y_sum = y1_values + y2_values

# Построение графиков
plt.figure(figsize=(10, 6))

# График f(x) = x^2
plt.plot(x_values, y1_values, label='f(x) = x^2', color='blue')

# График f(x) = 10*sin(x)
plt.plot(x_values, y2_values, label='f(x) = 10*sin(x)', color='green')

# График суммы
plt.plot(x_values, y_sum, label='Sum', color='red', linestyle='--')

# Настройка графика
plt.title('Graphs of Functions and Their Sum')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)

# Отображение графика
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Параметры сферы
R = 1
theta = np.linspace(0, np.pi / 2, 100)
phi = np.linspace(0, 2 * np.pi, 100)

# Параметризация сферы
theta, phi = np.meshgrid(theta, phi)
x = R * np.sin(theta) * np.cos(phi)
y = R * np.sin(theta) * np.sin(phi)
z = R * np.cos(theta)

# Построение 3D графика
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111, projection='3d')

# Построение верхней половины сферы
ax.plot_surface(x, y, z, color='blue', alpha=0.8)

# Настройка графика
ax.set_title('Upper Half of a Sphere')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.grid(True)

# Отображение графика
plt.show()

```

```python
import matplotlib.pyplot as plt
import numpy as np

# Задаем координаты вершин треугольника
x = np.array([0, 1, 0.5, 0])
y = np.array([0, 0, np.sqrt(3), 0])

# Построение графика с различными цветами и стилями линий
plt.figure(figsize=(6, 6))

# Линия между вершинами 1 и 2 (красная пунктирная линия)
plt.plot(x[0:2], y[0:2], color='red', linestyle='--', label='Side 1-2')

# Линия между вершинами 2 и 3 (зеленая штрих-пунктирная линия)
plt.plot(x[1:3], y[1:3], color='green', linestyle='-.', label='Side 2-3')

# Линия между вершинами 3 и 1 (синяя кривая линия)
plt.plot([x[0], x[2]], [y[0], y[2]], color='blue', linestyle='-', label='Side 1-3')

# Настройка графика
plt.title('Triangle with Different Colors and Line Styles')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True)
plt.legend()

# Отображение графика
plt.show()

```
